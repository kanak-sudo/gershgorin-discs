```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate A Random Matrix
A = np.random.uniform(-3,3,(5,5))+complex(0,1)*np.random.uniform(-3,3,(5,5))
```

```python
#Compute Centers, Radii, and EigenValues
centers = [A[i][i] for i in range(A.shape[0])]
radii = [sum(A[i])-A[i][i] for i in range(A.shape[0])]
evals = np.linalg.eigvals(A)
```

```python
#A Function To Plot The Discs
def circ(a,r):
    theta = np.linspace(0,2*3.14,1000)
    points = [r*np.exp(complex(0,1)*t)+a for t in theta]
    xvals = [z.real for z in points]
    yvals = [z.imag for z in points]
    return(xvals,yvals)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
for i in range(A.shape[0]):
    ax.plot(*circ(centers[i],radii[i]),label='GDisc of A'+str(i+1)+str(i+1))
ax.scatter(evals.real,evals.imag,s=100,marker='x',color='Black',label='Eigen Values')
ax.legend()
plt.show()
```
